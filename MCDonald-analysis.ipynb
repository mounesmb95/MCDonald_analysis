```python
from scipy.stats import chi2_contingency

# Sample DataFrame with columns 'Segment', 'Age', and 'Moral obligation'
data = {
    'Segment': [1, 2, 3, 4, 5, 6],
    'Age': [25, 30, 35, 40, 45, 50],  # Sample age values
    'Moral obligation': [3.5, 4.0, 3.8, 3.2, 4.2, 3.9]  # Sample moral obligation values
}

df = pd.DataFrame(data)

# Chi-square test for gender distribution across segments
# Since gender information is not provided, let's assume 'Gender' column doesn't exist
# Instead, we'll use 'Age' as the other variable for the chi-square test

observed = pd.crosstab(index=df['Segment'], columns=pd.cut(df['Age'], bins=3))
chi2, p, _, _ = chi2_contingency(observed)
print("Chi-square test for age distribution across segments:")
print("Chi-square:", chi2)
print("p-value:", p)
```

```python
from scipy.stats import kruskal

# Kruskal-Wallis H-test to test for differences in moral obligation values across segments
h_stat, p_val = kruskal(*[df[df['Segment'] == segment]['Moral_obligation'] for segment in df['Segment'].unique()])
print("\nKruskal-Wallis H-test for differences in moral obligation values across segments:")
print("H-statistic:", h_stat)
print("p-value:", p_val)
```
